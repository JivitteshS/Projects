```python
types=pd.DataFrame(df.dtypes)
objects=list()
num=list()
for i in range(0,14):
    if types.iloc[i,0] == "object":
        objects.append(df.columns[i])
    elif types.iloc[i,0] == "int64":
        num.append(df.columns[i])
for i in objects:
    c=pd.DataFrame(df[[i]].value_counts())
    print("The number of "+ i +" categories for  is: ",c.index.value_counts().sum())
```

```python
X=df_m
y=df_m["status"]
X=sm.add_constant(X)
```

```python
def get_tp_tn_fp_fn(actual, predicted):
    cm = confusion_matrix(actual, predicted)
    tp = cm[0, 0]
    tn = cm[1, 1]
    fp = cm[0, 1]
    fn = cm[1, 0]
    return tp, tn, fp, fn
```

## Baseline Model

```python
model=LogisticRegression()
model.fit(X,y)
pred_y=model.predict(X)
accuracy = np.sum(y == pred_y) / len(y)
tp, tn, fp, fn = get_tp_tn_fp_fn(y, pred_y)
recall=tp/(tp+fn)

print("Accuracy : "+str(round(accuracy,5)))
print("Recall: "+str(round(recall,5)))

model = sm.OLS(y, X).fit()
model.summary()
```

### Perform CHiSquare test of independence to check for effect of the categorical features on the chanes of defaule

```python
df
```

```python
df["status"]
```

```python
significant_cat_columns=list()
for i in df_cat.columns:
    observed_data = np.array(pd.crosstab(df_cat[i],df["status"]))
    chi2,p_value,dof, expected = chi2_contingency(observed_data)
    alpha=0.05
    if p_value<0.05:
        print(i+"has siginificant effect on the chances of default")
        significant_cat_columns.append(i)
```

### Preform one way anova to check for effect of continuous numerical columns on the status

```python
significant_num_columns=list()

for i in num:
    one=df[df["status"]==1]
    zero=df[df["status"]==0]
    a=one[i]
    b=zero[i]
    _,p_value=f_oneway(a,b)
    if p_value<0.05:
        print("there is significant differnce in average of "+i+" for the credit default status")
        significant_num_columns.append(i)
```

### Features which are significant after performing one way anova

```python
significant_num_columns
```

```python
df_num=df[significant_num_columns]
```

```python

X = df_num

# Calculate VIF for each independent variable
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

###### The VIF of the columns are more or less near to 5 so there isn't high multicolinearity among the continuous features
